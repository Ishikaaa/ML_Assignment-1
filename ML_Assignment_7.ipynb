```
#Q1.
import io
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
#importing dataset
df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv')
df
```

```
df.head()
```

```
#description about the dataset 
df.describe(include="all")
```

```
#Q2. Count of the missing values
df.isnull().sum()
```

```
# To handle missing values 
# i. If there is any missing value in row then full row will be removed
dfh1=df.dropna()
dfh1
```

```
#ii. remove only that rows in which values of all columns is missing 
dfh2=df.dropna(how="all")
dfh2
```

```
#iii. in this only that columns will remove in which values of all rows is missing 
dfh3=df.dropna(how="all",axis=1)
dfh3
```

```
#iv. by replacing missing values by 0 
df.fillna(0)
```

```
#v. by replacing missing value by mean of other values 
df['sqft_living15'].fillna(df['sqft_living15'].mean(),inplace=True)
df
```
