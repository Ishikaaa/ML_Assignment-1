```
# Install the PyDrive wrapper & import libraries.
# This only needs to be done once per notebook.
!pip install -U -q PyDrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once per notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

# Download a file based on its file ID.
#
# A file ID looks like: laggVyWshwcyP6kEI-y_W3P8D26sz
file_id = '1f92xSdbSedDbtZnD_Y5a_rbNG97kJL7v'
down = drive.CreateFile({'id': file_id})
```

```
#Q1.
import io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(io.StringIO(down.GetContentString()))
print("*1"+"*"*50+"Scatter Plot"+"*"*50)
sns.jointplot(x = 'MinTemp', y = 'Rainfall', data = df)
```

```
print("*2"+"*"*50+"Kernel Density Estimation"+"*"*50)
sns.jointplot(x='MinTemp', y='Rainfall', data=df, kind='kde')
```

```
print("*3"+"*"*50+"Kernel Density Estimation"+"*"*50)
sns.distplot(df['Rainfall'],bins=15)
```

```
print("*4"+"*"*50+"Histogram"+"*"*50)
sns.distplot(df['Rainfall'],bins=15, kde=False, rug=True)
```

```
print("*5"+"*"*50+"*"*50)
sns.kdeplot(df['Rainfall'], shade=True)
```

```
print("*6"+"*"*50+"*"*50)
sns.kdeplot(df['Rainfall'], label='original')
sns.kdeplot(df['Rainfall'], bw=1, label='bw=1')
sns.kdeplot(df['Rainfall'], bw=2, label='bw=2')
plt.legend()
```

```
print("*7"+"*"*100)
x=df['Rainfall'].T
y=df['MinTemp'].T
with sns.axes_style('white'):
    sns.jointplot(x = x, y = y, kind = 'hex', color = 'k')
```
