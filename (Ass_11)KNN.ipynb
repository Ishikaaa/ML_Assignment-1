# ***1. Import the dataset from sklearn***

```
import pandas as pd 
import numpy as np 
from sklearn.datasets import load_digits 
```

```
load_digits()
```

# ***2. Save X(features) and y(target) as dataframes -***

```
X=pd.DataFrame(load_digits().data)
X.head()
```

```
Y=pd.DataFrame(load_digits().target)
Y.head()
```

# ***3. Split the dataset into training and testing sets -***

```
# data preprocessing - As we will use eucleadian distance in KNN. And eucleadian distance gets oo much affected by scaling of features. So, here we will use 
# standardisation for feature scaling. 
from sklearn.preprocessing import StandardScaler 
stX=StandardScaler().fit_transform(X)
stY=StandardScaler().fit_transform(Y)
```

```
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
X_train,X_test,Y_train,Y_test
```

# ***4. Use KNN - ***

```
from sklearn.neighbors import KNeighborsClassifier 
classifier=KNeighborsClassifier()
classifier.fit(X_train,Y_train)
```

```
# Predicting Accuracy 
Y_pred=classifier.predict(X_test)
```

# ***5. KNN Accuracy Score - ***

```
from sklearn import metrics 
metrics.accuracy_score(Y_test,Y_pred)
```
