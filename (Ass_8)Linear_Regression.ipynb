# ***1. Creating Working Directory - *

```
# importing libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```
dataset=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv ")
dataset.head()
```

```
dataset.shape
```

```
dataset.columns
```

```
dataset.info() 
```

**Note - **

1.   As number of rows is equal to count. So, there is no missing value. 



# ***2. Feature Extraction - ***

**Note - **


1.  Only 'Date' column is object. But we don't need ''Date' column to predict price. SO we can directly remove that column instead of converting categorical feature into numerical feature. 
2. 'Id' colunm is also of no use. So we can also remove that column. 



```
dataset.drop(['date'],inplace=True,axis=1)
dataset.head()
```

```
dataset.drop(['id'],inplace=True,axis=1)
dataset.columns
```

```
# standardisation
from sklearn.preprocessing import StandardScaler 
StandardScaler().fit_transform(dataset)
```

# ***2. Splitting dataset into training and test set --***

```
X=dataset[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']]
Y=dataset['price']
X.head()
```

```
Y.head()
```

```
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
X_train,X_test,Y_train,Y_test
```

# ***3. Training - ***

```
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

# ***4. Making The Predictions - ***

```
y_pred=regressor.predict(X_test)
```

```
compdata=pd.DataFrame({"Actual Data":Y_test,"Predicted Data":y_pred})
compdata
```

```
compdata['Predicted Data']=compdata['Predicted Data'].astype(float)
compdata
```

# ***5. Evaluating The Algorithm - ***

```
from sklearn import metrics 
print("Mean Absolute Error:",metrics.mean_absolute_error(Y_test,y_pred))
print("Mean Squared Error:",metrics.mean_squared_error(Y_test,y_pred))
print("Root Mean Squared Error:",np.sqrt(metrics.mean_squared_error(Y_test,y_pred)))
```
